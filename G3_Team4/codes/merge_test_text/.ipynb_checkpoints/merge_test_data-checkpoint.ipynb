# Merge LDA , k means , textblov and vader for test set 

```python
## Import packages
import numpy as np
import pandas as pd
import copy
```

```python
kmeans = pd.read_csv('kmeans_test_text.csv')
kmeans.head(5)
```

```python
lda = pd.read_csv('lda_test_text_5.csv')
lda.head(5)
```

```python
testVader = pd.read_csv('testVader.csv')
testVader.head(5)
```

```python
testTextblob = pd.read_csv('testTextblob.csv')
testTextblob.head(5)
```

```python
print(kmeans.shape)
print(lda.shape)
print(testVader.shape)
print(testTextblob.shape)
```

```python
def Sentiscore(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Postive'
```

```python
merged_test_data = pd.DataFrame(kmeans,columns=['id', 'title' , 'text' , 'K_means'])
merged_test_data['LDA_Dominant_Topic'] = lda['Dominant_Topic']
merged_test_data['text_vader_compound'] = testVader['compound']
merged_test_data['text_vader_neg'] = testVader['neg']
merged_test_data['text_vader_neu'] = testVader['neu']
merged_test_data['text_vader_pos'] = testVader['pos']
merged_test_data['tb_Pol'] = testTextblob['tb_Pol']
merged_test_data['tb_Subj'] = testTextblob['tb_Subj']
merged_test_data['Polarity_category'] = testTextblob['tb_Pol'].apply(Sentiscore)
```
